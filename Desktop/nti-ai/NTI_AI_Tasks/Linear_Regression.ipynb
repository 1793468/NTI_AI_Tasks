```python
df = df.copy()
columns = ['Years of Experience', 'Salary']

plt.figure(figsize=(12, 6))
for i, col in enumerate(columns, 1):
    plt.subplot(1, 2, i)
    sns.boxplot(df[col])
    plt.title(f'Box Plot of {col}')

plt.tight_layout()
plt.show()
```

```python
df=df.copy()
for col in columns:
    if col in df.columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        df[col + '_outlier'] = df[col].apply(lambda x: x < lower_bound or x > upper_bound)
        print(f"Number of outliers in {col}: {df[col + '_outlier'].sum()}")
```

```python
from scipy import stats

for col in columns:
    if col in df.columns:
        df[col + '_zscore'] = stats.zscore(df[col].dropna())
        df[col + '_outlier'] = df[col + '_zscore'].apply(lambda x: abs(x) > 3)
        print(f"Number of outliers in {col}: {df[col + '_outlier'].sum()}")
```

```python
df=df.copy()
plt.figure(figsize=(12, 6))
for i, col in enumerate(columns, 1):
    plt.subplot(1, 2, i)
    sns.histplot(df[col + '_zscore'], bins=30, kde=True)
    plt.title(f'Z-Score Distribution of {col}')
    plt.xlabel('Z-Score')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
df=df.copy()

df_cleaned = df[~df[[col + '_outlier' for col in columns if col in df.columns]].any(axis=1)]

print(f"Number of rows after dropping outliers: {df_cleaned.shape[0]}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
df = df.copy()
class LinearRegressionGD:
    def __init__(self, learning_rate=0.01, n_iterations=1000):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.theta0 = 0
        self.theta1 = 0

    def fit(self, X, Y):
        X = np.array(X)
        Y = np.array(Y)
        self.theta0, self.theta1 = np.random.rand(2)  # Initialize parameters

        for _ in range(self.n_iterations):
            y_predicted = self.theta0 + self.theta1 * X
            errors = Y - y_predicted

            # Update parameters using gradient descent
            self.theta0 -= self.learning_rate * -np.mean(errors)
            self.theta1 -= self.learning_rate * -np.mean(errors * X)

    def predict(self, X):
        X = np.array(X)
        return self.theta0 + self.theta1 * X


```

```python

np.random.seed(42)
X = 2 * np.random.rand(100, 1).flatten()
y = 4 + 3 * X + np.random.randn(100)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python

model = LinearRegressionGD(learning_rate=0.01, n_iterations=1000)
model.fit(X_train, y_train)
```

```python

y_predicted = model.predict(X_test)

print("Slope: ", model.theta1)
print("Y-Intercept: ", model.theta0)
print('Mean Squared Error: ', mean_squared_error(y_test, y_predicted))
print("R² Score: ", r2_score(y_test, y_predicted))
```

```python

plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.scatter(X_test, y_test, color='green', label='Test Data')
plt.plot(X_test, y_predicted, color='red', linewidth=2, label='Regression Line')
plt.title('Linear Regression using Gradient Descent')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

```

```python
import seaborn as sns

# Predict values for plotting confidence intervals
X_plot = np.linspace(X.min(), X.max(), 100)
y_plot = model.predict(X_plot)

plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X_plot, y_plot, color='red', linewidth=2, label='Regression Line')

# Plot confidence intervals
residuals = y - model.predict(X)
residual_std = np.std(residuals)
plt.fill_between(X_plot, y_plot - 1.96 * residual_std, y_plot + 1.96 * residual_std, color='red', alpha=0.2)

plt.title('Regression Line with Confidence Intervals')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
train_errors, val_errors = [], []
for m in range(1, len(X_train)):
    model.fit(X_train[:m], y_train[:m])
    y_train_predict = model.predict(X_train[:m])
    y_val_predict = model.predict(X_test)
    train_errors.append(mean_squared_error(y_train[:m], y_train_predict))
    val_errors.append(mean_squared_error(y_test, y_val_predict))

plt.figure(figsize=(10, 6))
plt.plot(np.sqrt(train_errors), label='Train Error')
plt.plot(np.sqrt(val_errors), label='Validation Error')
plt.title('Learning Curve')
plt.xlabel('Number of Training Instances')
plt.ylabel('RMSE')
plt.legend()
plt.show()
```
