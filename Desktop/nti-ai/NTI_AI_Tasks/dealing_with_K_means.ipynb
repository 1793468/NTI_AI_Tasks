```python
columns = ['ash', 'alcalinity_of_ash', 'magnesium','proanthocyanins',
           'color_intensity', 'hue']
for _ in range(2):  # Repeat the process 3 times
    for col in columns:
        if col in df.columns:
            # Calculate IQR for the column
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            # Mark outliers
            df[col + '_outlier'] = df[col].apply(lambda x: x < lower_bound or x > upper_bound)

            # Count and print the number of outliers
            print(f"Number of outliers in {col}: {df[col + '_outlier'].sum()}")

    # Create a copy of the DataFrame excluding rows with outliers
    df_cleaned = df[~df[[col + '_outlier' for col in columns if col in df.columns]].any(axis=1)].copy()

    # Update the original DataFrame for the next iteration
    df = df_cleaned.copy()

    # Print the number of rows after each iteration
    print(f"Number of rows after dropping outliers (round {_ + 1}): {df_cleaned.shape[0]}")

# Final cleaned DataFrame after 3 iterations
df_cleaned = df.copy()
```

```python
covariance_matrix = np.cov(df_scaled, rowvar=False)

print("Covariance Matrix:")
print(covariance_matrix)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)
```

```python
sorted_indices = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_indices]
sorted_eigenvectors = eigenvectors[:, sorted_indices]
```

```python
explained_variance_ratio = sorted_eigenvalues / np.sum(sorted_eigenvalues)
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)



plt.figure(figsize=(10, 5))
plt.plot(cumulative_variance_ratio, marker='o')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid()
plt.axhline(y=0.95, color='r', linestyle='--')  # Example threshold
plt.show()
```

**K-means , PCA from Scratch**

```python
# Initialize centroids for K-means
def init_centroids(df_scaled, k, random_state=None):
    """
    Initialize k centroids from the dataset df_scaled.

    Parameters:
    - df_scaled: Scaled dataset (m x n) where m is the number of examples and n is the number of features.
    - k: Number of clusters.
    - random_state: Optional; Set the random seed for reproducibility.

    Returns:
    - centroids: Initial k centroids chosen randomly from the dataset df_scaled.
    """
    m, n = df_scaled.shape
    centroids = np.zeros((k, n))

    # Set random seed if provided for reproducibility
    if random_state is not None:
        np.random.seed(random_state)

    idx = np.random.randint(0, m, k)
    for i in range(k):
        centroids[i, :] = df_scaled[idx[i], :]

    return centroids
print(centroids.shape)
```

```python
# Find closest centroids for each point
def find_closest_centroids(df_scaled, centroids):
    """
    Assign each data point to the closest centroid.

    Parameters:
    - df_scaled: Scaled dataset (m x n) where m is the number of examples and n is the number of features.
    - centroids: Current centroids (k x n), where k is the number of centroids.

    Returns:
    - idx: An array of centroid assignments for each example in df_scaled.
    """
    m = df_scaled.shape[0]
    k = centroids.shape[0]
    idx = np.zeros(m)

    for i in range(m):
        min_dist = float('inf')
        for j in range(k):
            dist = np.sum((df_scaled[i, :] - centroids[j, :]) ** 2)
            if dist < min_dist:
                min_dist = dist
                idx[i] = j

    return idx
```

```python
def compute_centroids(df_scaled_np, idx, k):
    """
    Recompute centroids as the mean of all points assigned to each centroid.

    Parameters:
    - df_scaled_np: Scaled dataset as a NumPy array (m x n).
    - idx: Centroid assignments for each example.
    - k: Number of centroids.

    Returns:
    - centroids: Updated centroids (k x n).
    """
    m, n = df_scaled_np.shape
    centroids = np.zeros((k, n))

    for i in range(k):
        # Find all points assigned to centroid i
        indices = np.where(idx == i)[0]
        if len(indices) > 0:  # Avoid division by zero if no points are assigned to a centroid
            # Calculate the mean of points assigned to centroid i
            centroids[i, :] = np.mean(df_scaled_np[indices], axis=0)

    return centroids
```

```python
 # Full K-means function
def run_k_means(df_scaled, initial_centroids, max_iters=10):
    """
    Run the K-means clustering algorithm.

    Parameters:
    - df_scaled: Scaled dataset (m x n).
    - initial_centroids: Initial centroids (k x n).
    - max_iters: Maximum number of iterations for K-means.

    Returns:
    - idx: Final centroid assignments for each example.
    - centroids: Final computed centroids.
    """
    m, n = df_scaled.shape
    k = initial_centroids.shape[0]
    idx = np.zeros(m)
    centroids = initial_centroids

    for i in range(max_iters):
        idx = find_closest_centroids(df_scaled, centroids)
        centroids = compute_centroids(df_scaled, idx, k)

    return idx, centroids
```

```python
# PCA Implementation
def pca(df_scaled):
    """
    Perform Principal Component Analysis (PCA) on the dataset df_scaled.

    Parameters:
    - df_scaled: Scaled dataset (m x n).

    Returns:
    - U: Left singular vectors (n x n) where n is the number of features.
    - S: Singular values (n).
    - V: Right singular vectors (n x n).
    """
    # Normalize the features
    df_scaled = (df_scaled - df_scaled.mean(axis=0)) / df_scaled.std(axis=0)

    # Compute the covariance matrix
    df_scaled = np.matrix(df_scaled)
    cov = (df_scaled.T * df_scaled) / df_scaled.shape[0]

    print('Covariance matrix:\n', cov)

    # Perform Singular Value Decomposition (SVD)
    U, S, V = np.linalg.svd(cov)

    return U, S, V
```

```python
# Project data onto k principal components
def project_data(df_scaled, U, k):
    """
    Project the dataset df_scaled onto the top k principal components.

    Parameters:
    - df_scaled: Scaled dataset (m x n).
    - U: Matrix of eigenvectors (n x n) from PCA.
    - k: Number of principal components to project onto.

    Returns:
    - Z: The projection of df_scaled onto the first k principal components.
    """
    U_reduced = U[:, :k]
    return np.dot(df_scaled, U_reduced)
```

```python
# Recover data from the reduced projection
def recover_data(Z, U, k):
    """
    Recover the data from its projection onto the k principal components.

    Parameters:
    - Z: The projected data (m x k).
    - U: Matrix of eigenvectors (n x n) from PCA.
    - k: Number of principal components.

    Returns:
    - df_rec: Approximation of the original data (m x n) from the projection.
    """
    U_reduced = U[:, :k]
    return np.dot(Z, U_reduced.T)
```

```python
import numpy as np

# Ensure df_scaled is converted to a NumPy array for proper indexing
df_scaled_np = df_scaled.to_numpy()  # Convert Pandas DataFrame to NumPy array

# Initialize centroids for K-means
k = 3  # Number of clusters
random_state = 42  # To ensure reproducibility

# Initialize centroids and run K-means clustering
initial_centroids = init_centroids(df_scaled_np, k=k, random_state=random_state)
idx, centroids = run_k_means(df_scaled_np, initial_centroids, max_iters=10)

# K-means output
print("Centroid Assignments (idx):", idx)
print("Final Centroids:", centroids)

# Now run PCA on the scaled dataset
U, S, V = pca(df_scaled_np)

# Project the data onto the first 2 principal components
k_components = 2  # Number of principal components to project onto
Z = project_data(df_scaled_np, U, k=k_components)

# PCA output
print("Projected Data (Z):", Z)

# Recover the data from the projection
df_rec = recover_data(Z, U, k=k_components)

# Recovered data output
print("Recovered Data from Projection (df_rec):", df_rec)
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.decomposition import PCA
# ====================
# K-MEANS EVALUATION
# ====================

# Inertia (Within-Cluster Sum of Squares)
def compute_inertia(df_scaled_np, centroids, idx):
    inertia = 0
    for i in range(len(df_scaled_np)):
        inertia += np.sum((df_scaled_np[i] - centroids[int(idx[i])])**2)
    return inertia

inertia = compute_inertia(df_scaled_np, centroids, idx)
print("K-means Inertia (Within-Cluster Sum of Squares):", inertia)

# Silhouette Score
sil_score = silhouette_score(df_scaled_np, idx)
print("Silhouette Score:", sil_score)

# Davies-Bouldin Index
db_index = davies_bouldin_score(df_scaled_np, idx)
print("Davies-Bouldin Index:", db_index)

# ====================
# K-MEANS VISUALIZATION
# ====================

# Visualizing Clusters (2D)
def visualize_kmeans(df_scaled_np, idx, centroids):
    plt.scatter(df_scaled_np[:, 0], df_scaled_np[:, 1], c=idx, s=50, cmap='viridis')
    plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, alpha=0.75, marker='X')
    plt.title("K-means Clustering (2D visualization)")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()

# Plot K-means Clusters
visualize_kmeans(df_scaled_np, idx, centroids)

# ====================
# ELBOW METHOD for K-MEANS
# ====================
def elbow_method(df_scaled_np, max_k=10):
    inertias = []
    for k in range(1, max_k+1):
        initial_centroids = init_centroids(df_scaled_np, k=k, random_state=random_state)
        idx, centroids = run_k_means(df_scaled_np, initial_centroids, max_iters=10)
        inertia = compute_inertia(df_scaled_np, centroids, idx)
        inertias.append(inertia)
    plt.plot(range(1, max_k+1), inertias, marker='o')
    plt.title('Elbow Method')
    plt.xlabel('Number of clusters (k)')
    plt.ylabel('Inertia (Sum of Squared Distances)')
    plt.show()

# Plot Elbow Method
elbow_method(df_scaled_np, max_k=10)

# ====================
# PCA VISUALIZATION & EVALUATION
# ====================

# Run PCA
pca = PCA()
pca.fit(df_scaled_np)
explained_variance = pca.explained_variance_ratio_

# Plot Explained Variance Ratio
def plot_explained_variance(explained_variance):
    plt.plot(np.cumsum(explained_variance), marker='o')
    plt.title("PCA Explained Variance Ratio")
    plt.xlabel("Number of Components")
    plt.ylabel("Cumulative Explained Variance")
    plt.show()

# Plot PCA Explained Variance
plot_explained_variance(explained_variance)

# Project data onto the first 2 principal components
k_components = 2  # Number of principal components to project onto
Z = pca.transform(df_scaled_np)[:, :k_components]

# Visualizing PCA (first 2 principal components)
def visualize_pca(Z):
    plt.scatter(Z[:, 0], Z[:, 1], s=50, cmap='viridis')
    plt.title("PCA Projection (2D)")
    plt.xlabel("Principal Component 1")
    plt.ylabel("Principal Component 2")
    plt.show()

# Plot PCA Projection
visualize_pca(Z)

# Reconstruction Error
def reconstruction_error(df_scaled_np, Z, k_components=2):
    """
    Compute the reconstruction error by projecting onto the first k principal components
    and reconstructing the data.

    Parameters:
    - df_scaled_np: Original scaled dataset (m x n).
    - Z: Projected data (m x k).
    - k_components: Number of principal components to use for reconstruction.

    Returns:
    - mse: Mean squared reconstruction error.
    """
    # Create a new PCA object with the required number of components
    pca_reduced = PCA(n_components=k_components)
    # Fit the PCA object to the original data
    pca_reduced.fit(df_scaled_np)
    # Use the PCA object to inverse transform the reduced data Z
    df_rec = pca_reduced.inverse_transform(Z)
    # Compute mean squared error between the original and reconstructed data
    mse = np.mean((df_scaled_np - df_rec) ** 2)
    return mse

# Calculate Reconstruction Error
reconstruction_err = reconstruction_error(df_scaled_np, Z, k_components=2)
print("PCA Reconstruction Error:", reconstruction_err)
```
