```python
columns = ['age', 'education', 'cigsPerDay', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']  # Specify your columns

for _ in range(6):  # Repeat the process 3 times
    for col in columns:
        if col in df.columns:
            # Calculate IQR for the column
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            # Mark outliers
            df[col + '_outlier'] = df[col].apply(lambda x: x < lower_bound or x > upper_bound)

            # Count and print the number of outliers
            print(f"Number of outliers in {col}: {df[col + '_outlier'].sum()}")

    # Create a copy of the DataFrame excluding rows with outliers
    df_cleaned = df[~df[[col + '_outlier' for col in columns if col in df.columns]].any(axis=1)].copy()

    # Update the original DataFrame for the next iteration
    df = df_cleaned.copy()

    # Print the number of rows after each iteration
    print(f"Number of rows after dropping outliers (round {_ + 1}): {df_cleaned.shape[0]}")

# Final cleaned DataFrame after 3 iterations
df_cleaned = df.copy()
```

```python
import numpy as np

class LogisticRegression:
    def __init__(self):
        self.Thetas = None
        self.cost_history = []  # To store the cost function values over iterations

    def sigmoid(self, Thetas, x):
        z = np.dot(x, Thetas)  # Dot product between X and Thetas
        return 1 / (1 + np.exp(-z))  # Sigmoid function

    def gradient_descent(self, X, y, Thetas, learning_rate, iterations):
        m = len(y)
        for i in range(iterations):
            h = self.sigmoid(Thetas, X)  # Get the predicted values (hypothesis)
            gradient = (1 / m) * np.dot(X.T, (h - y))  # Calculate gradient
            Thetas = Thetas - learning_rate * gradient  # Update Thetas
            cost = self.cost(y, h)  # Calculate and store cost
            self.cost_history.append(cost)  # Save the cost for plotting
        return Thetas

    def cost(self, y, h):
        m = len(y)
        cost = -(1 / m) * np.sum(y * np.log(h) + (1 - y) * np.log(1 - h))  # Log-loss function
        return cost

# Prepare data
X = np.array(df[['male', 'age',  'currentSmoker', 'cigsPerDay', 'prevalentHyp',  'totChol', 'sysBP','prevalentStroke','BPMeds' ]])
X = np.c_[np.ones((X.shape[0], 1)), X]  # Add column of ones for the bias term
y = np.array(df[['TenYearCHD']]).reshape(-1, 1)  # Ensure y is a 2D array

Thetas = np.zeros((X.shape[1], 1))  # Initialize Thetas with the shape based on features

# Parameters
learning_rate = 0.1
iterations = 1200

# Logistic regression model
model = LogisticRegression()

# Gradient descent optimization
Thetas = model.gradient_descent(X, y, Thetas, learning_rate, iterations)

# Sigmoid function output with final Thetas
h = model.sigmoid(Thetas, X)

# Compute final cost
cost_value = model.cost(y, h)

# Output
print(f"Final cost: {cost_value}")
print(f"Final Thetas: {Thetas}")
#######################################################################################################################
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, confusion_matrix

# Convert sigmoid outputs (probabilities) into binary predictions (0 or 1)
y_pred_prob = model.sigmoid(Thetas, X)  # Probabilities
y_pred = (y_pred_prob >= 0.5).astype(int)  # Convert probabilities to binary values (0 or 1)

# Calculate evaluation metrics
accuracy = accuracy_score(y, y_pred)
precision = precision_score(y, y_pred)
recall = recall_score(y, y_pred)
auc = roc_auc_score(y, y_pred_prob)

# Confusion matrix
conf_matrix = confusion_matrix(y, y_pred)

# Print metrics
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"AUC: {auc}")
print(f"Confusion Matrix:\n{conf_matrix}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression as SklearnLogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, confusion_matrix, roc_curve, precision_recall_curve
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your DataFrame
X = np.array(df[['male', 'age', 'currentSmoker', 'cigsPerDay', 'prevalentHyp', 'totChol', 'sysBP', 'prevalentStroke', 'BPMeds']])
y = np.array(df['TenYearCHD'])

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Initialize and train the logistic regression model
sklearn_model = SklearnLogisticRegression(max_iter=1200, solver='lbfgs')
sklearn_model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred_prob = sklearn_model.predict_proba(X_test)[:, 1]  # Probabilities for the positive class
y_pred = sklearn_model.predict(X_test)  # Binary predictions

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred_prob)

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Print metrics
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"AUC: {auc}")
print(f"Confusion Matrix:\n{conf_matrix}")
```

```python
# ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
plt.plot(fpr, tpr, color='blue', label='ROC Curve')
plt.plot([0, 1], [0, 1], color='red', linestyle='--', label='Random Guessing')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
# Confusion Matrix Heatmap

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False, xticklabels=['No Disease', 'Disease'], yticklabels=['No Disease', 'Disease'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Precision-Recall Curve
precision, recall, _ = precision_recall_curve(y_test, y_pred_prob)
plt.plot(recall, precision, marker='.', label='Precision-Recall Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.show()
```
